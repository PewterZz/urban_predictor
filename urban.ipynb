```python
data.replace("-", float("nan"), inplace=True)

# Drop rows with NaN values
data.dropna(inplace=True)

# Reset the index
data.reset_index(drop=True, inplace=True)

data
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Create a DataFrame with the provided data
data = pd.DataFrame({
    'Year': range(2005, 2017),
    'Pond Area': [512524.0, 486982.0, 452901.0, 613175.0, 682725.0, 682857.0, 749220.0, 657346.0, 650509.3084, 667083.0, 715846, 674135]
})

# Set the 'Year' column as the index
data.set_index('Year', inplace=True)

# Set the range of future years for prediction
start_year = 2017
end_year = 2022
future_years = range(start_year, end_year + 1)

# Create the ARIMA model
model = ARIMA(data, order=(1, 1, 1))
model_fit = model.fit()

# Make predictions for the future years
predictions = model_fit.predict(start=len(data), end=len(data) + len(future_years) - 1)

# Get the actual data for 2022
actual_data_2022 = 785000

# Calculate the total predicted area for the future years
total_predicted_area = np.sum(predictions)

# Print the predicted area for each future year
for year, prediction in zip(future_years, predictions):
    print(f"Year: {year}, Predicted Area: {prediction}")

# Print the actual data for 2022
print(f"Actual Area (2022): {actual_data_2022}")

# Print the total predicted area for the future years
print(f"Total Predicted Area ({start_year}-{end_year}): {total_predicted_area}")

```
